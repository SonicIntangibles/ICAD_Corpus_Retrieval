```python
# Count and plot the number of files in a set of subfolders
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from collections import Counter

# Specify root folder and enumerate all subfolders
root_folder = '/content/drive/MyDrive/ICAD_Papers'
subfolders = [f.path for f in os.scandir(root_folder) if f.is_dir()]

# Count files in each subfolder
file_counts = {}
for subfolder in subfolders:
    num_files = len([f for f in os.listdir(subfolder) if os.path.isfile(os.path.join(subfolder, f))])
    file_counts[os.path.basename(subfolder)] = num_files
    print(f"Folder '{os.path.basename(subfolder)}' contains {num_files} files.")


# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(file_counts.keys(), file_counts.values())
plt.xlabel("Subfolders")
plt.ylabel("Number of Files")
plt.title("File Count in Subfolders")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
